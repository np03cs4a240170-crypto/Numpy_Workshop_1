```python
import pandas as pd
data = {
    "Name": ["santusti", "aanchal", "diya", "eluja", "rubina", "isha", None],
    "Age": [20, 21, None, 24, 22, 24, 23],
    "Gender": ["F", "f", "f", None, "F", "M", "F"],
    "Marks": [88, 92, 85, 90, None, 95, 88]
}
df = pd.DataFrame(data)
df.to_csv("medical_student.csv", index=False)

print("✅ medical_student.csv created successfully!")
```

```python
df.head()
```

```python
df.info()
print("missing values per column:")
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df['Sex'] = ['male', 'female', 'male', 'female', 'male', 'female']
df['Survived'] = [0, 1, 0, 1, 0, 1]
df.head()
```

```python

survival_by_sex = df.groupby('Sex')['Survived'].mean()
print("Mean survival rates by gender:")
print(survival_by_sex)
```

```python
Sex
female    1.0
male      0.0
Name: Survived, dtype: float64
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x=survival_by_sex.index, y=survival_by_sex.values)
plt.ylabel("Mean Survival Rate")
plt.title("Mean Survival Rate by Gender")
plt.show()
```

```python

sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Counts by Gender")
plt.show()
```

```python
import pandas as pd
data = {
    "Name": ["Nigima", "Aaditya", "Saurav", "Futrelle", "Harper", "Icard"],
    "Pclass": [1, 1, 2, 1, 3, 1],
    "Sex": ["male", "female", "male", "female", "male", "female"],
    "Age": [35, 28, 22, 32, 19, 24],
    "Fare": [120.0, 150.5, 30.0, 200.0, 15.0, 180.0],
    "Survived": [1, 1, 0, 1, 0, 1],
    "Embarked": ['C', 'Q', 'S', 'C', 'S', 'C']
}

df = pd.DataFrame(data)
df.head()
```

```python

survival_by_embarked = df.groupby('Embarked')['Survived'].mean()
print("Mean survival rate by port of embarkation:")
print(survival_by_embarked)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x=survival_by_embarked.index, y=survival_by_embarked.values, palette="Set2")
plt.ylabel("Mean Survival Rate")
plt.xlabel("Port of Embarkation")
plt.title("Mean Survival Rate by Embarkation Port")
plt.show()
```

```python

sns.catplot(x='Embarked', y='Survived', hue='Sex', kind='bar', data=df, palette="Set1")
plt.title("Survival Rate by Embarkation Port and Gender")
plt.show()
```
